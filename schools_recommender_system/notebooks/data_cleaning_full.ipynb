```python
with open("full_data.json", encoding='utf-8') as f:
    data = json.load(f)
    df = pd.DataFrame(data)
    
df1 = df.dropna()
```

```python
# Function to split Arabic and English names
def split_names(name):
    parts = name.split(' - ')
    if len(parts) >= 2:
        return parts[0], parts[1]
    else:
        return name, None


# Applying the function to the 'name' column and creating new columns for Arabic and English names
df1[['arabic_name', 'school_name']] = df1['name'].apply(lambda x: pd.Series(split_names(x)))
```
